## What Makes a Player Expensive?

## Infographic Final Product
![What Makes a Player Expensive](what_makes_a_player_expensive.png)

The final project is an infographic that explains my thought process on how to pinpoint "What makes a player expensive". The infographic includes the KDE graph of the spread of players' market value represented in a log format for readability. Then includes a correlation matrix graph to find what attributes strongly correlate to a player's market value. Four graphs that display the split of players' age in the respective positions (LW, RB, CM, and ST) of high-value players (Overall 80+). Under the four graphs, it displays a dataframe of how many young players, old players, and the ratio of young players to total players are in those positions. Lastly, it includes a takeaway. To explain the proper contents and explanations of the infographic, out of all the graphs so far, the heatmap was the most conclusive in answering "**What makes a player expensive**". The confusion matrix displays the relationship between two attributes. As it displayed, **"overall" and "potential" are the most determinative of a player's value**. This tracks due to the fact that a player with a higher overall means they are "better overall". A player who is "better" is valued higher. Secondly, a player's potential highly influences their market value due to "potential gain" in terms of future "bank for buck". In addition to that, the strongest correlation between a player's attributes is their "dribbling" and "passing". This means that if a player is good at either "dribbling" or "passing," they are most likely just as well-versed in the other. Dribbling makes sense due to the fact that "ball control" falls under this attribute. With poor "ball control", it in turn affects the other attributes as well. Better ball control means faster dribbling (pace), better set-up touch for shooting, and better understanding of hitting the ball at angles (passing). But, a player's "dribbling" is the most influential trait that affects their other attributes (excluding defending and physical). This is surprising since I would have expected that their market value affects their other attributes. This proves that a well-rounded player isn't necessarily a high-value player. In addition to that, the attribute that least affects a player's market value is their defensive skill, which is misleading and, in actuality, incorrect since most of the higher value players are "offensive-based", not "defensive".
