#### Checking for null values

Since I am planning to incorporate specific columns, it is best that I check for any null values in any columns for any observations

```python
import pandas as pd

# original dataframe
df = pd.read_csv("FC26_20250921.csv")
df.head()
```

```python
cols_to_check = [
    "player_id",
    "short_name",
    "age",
    "nationality_name",
    "player_positions",
    "overall",
    "value_eur",
    "pace",
    "shooting",
    "passing",
    "dribbling",
    "defending",
    "physic"
]

df[cols_to_check].isna().sum().sort_values(ascending = False)
```

There are only missing values in the in-game attribute columns. An equal amount of missing values, therefore, there must be a reason they are missing and to continue I must know why.

```python
rows_with_nulls = df[df[cols_to_check].isna().any(axis = 1)]
rows_with_nulls
```

After expanding on the missing in-game attributes, all of the players who are missing these values are goalkeepers, which makes sense. In order to further explore the relationship and reach a conclusion, the incorporation of goalkeepers isn't ideal, since it would skew the data. With this in mind, I will be exploring all players, excluding the goalkeepers.

### New Dataframe

The following dataframe only includes in-field players and the specific columns needed

```python
# columns needed
cols_needed = [
    "player_id",
    "short_name",
    "age",
    "nationality_name",
    "player_positions",
    "overall",
    "potential",
    "value_eur",
    "pace",
    "shooting",
    "passing",
    "dribbling",
    "defending",
    "physic"
]

# removes goalkeepers
df_no_gk = df[~df["player_positions"].str.contains("GK", na = False)]

# filters the dataframe
df_filtered = df_no_gk[cols_needed]
df_filtered.head()
```

Since I am planning to compare players by position, it is best that each player only represents their primary position

```python
df_filtered.loc[:, "player_positions"] = (
    df_filtered["player_positions"]
    .str.split(",")
    .str[0]
    .str.strip()
)
df_filtered
```
